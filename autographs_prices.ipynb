```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
import html5lib

from time import sleep
```

```python
autographs_prices = {}

for page in range(1,85):
    res = requests.get('https://steamcommunity.com/market/search?appid=570&q=%D0%B0%D0%B2%D1%82%D0%BE%D0%B3%D1%80%D0%B0%D1%84#p{page}_default_desc'.format(page=page))
    tree = BeautifulSoup(res.text, 'html5lib')
    
    # name
    for name_raw in tree.find_all(class_="market_listing_item_name"):
        name = name_raw.text.replace('Автограф:','')
        autographs_prices.update({'aut_author': name}) 
            
    # num of autographs
    for num_raw in tree.find_all(class_="market_listing_num_listings_qty"):
        num = num_raw.text
        autogrphs_prices.update({'aut_count':num})
    
    # price
    for price_raw in tree.find_all(class_="normal_price"):
        price = float(price_raw.text.replace(' USD','').replace('$',''))
        autographs_prices.update({'aut_price_usd':price})
        
    sleep(2)
```
