## Getting all autograph prices

```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
import html5lib
import json

from time import sleep
```

```python
prices_all_dict = pd.DataFrame()

for page in range(0,6036,100):
    res = requests.get('https://steamcommunity.com/market/search/render/?query=autograph&norender=1&count=500&start={page}'.format(page=page))
    prices_dict = json.loads(res.text)
    prices_df = pd.DataFrame(prices_dict['results'])
    prices_all_dict = pd.concat([prices_df, prices_all_dict], ignore_index=True)
    
    sleep(50)
```

```python
subscription_key = '269b18656960431c9cf7a2f2a943e15c' # active only for 7 days, fella
```

```python
search_url = "https://api.cognitive.microsoft.com/bing/v7.0/search"
```

```python
player_search_count = {}

for player in players_list:
    headers = {"Ocp-Apim-Subscription-Key" : subscription_key}
    params  = {"q": "{player} dota 2".format(player=player), 
               "textDecorations":True, 
               "textFormat":"HTML"}
    response = requests.get(search_url, headers=headers, params=params)
    response.raise_for_status()
    search_results = response.json()
    if search_results['webPages']['totalEstimatedMatches'] is not None:
        count_search = search_results['webPages']['totalEstimatedMatches']
    else:
        count_search = 0
    player_search_count.update({player:count_search})
    
    sleep(2)
```

```python
players_df_steam = pd.DataFrame.from_dict(player_search_count, orient='index').reset_index()
```

```python
players_df_steam_all = pd.merge(prices_all_dict, players_df_steam, left_on='name', right_on='index', how='left')
```

```python
import numpy as np
df = players_df_steam_all[['sell_price_text', 0]]
df['sell_price_text'] = df['sell_price_text'].str.replace('$','').str.replace(',','').astype(float)
df['sell_price_text'] = np.log(df['sell_price_text'])
df[0] = np.log(df[0])
```
